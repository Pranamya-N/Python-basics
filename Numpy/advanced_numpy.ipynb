```python
# Broadcasting is the process of doing the arithmetic operation of the numpy arrays of the different dimensions
```

RULES OF BROADCASTING

# Here’s how broadcasting works for your specific arrays:

Array 1 has a shape of (3, 4).
Array 2 has a shape of (4), which is equivalent to (1, 4).
Step-by-step Broadcasting Rules:
Align shapes from the right:

Array 1: (3, 4)
Array 2: (1, 4) (NumPy automatically prepends a 1 to the shape of the second array to align dimensions.)
Check compatibility:

The rightmost dimension matches: 4 == 4 (compatible).
The second-to-last dimension is 3 in Array 1 and 1 in Array 2: Since 1 is compatible with any dimension, it can be "stretched."
Thus, broadcasting is possible, and NumPy will behave as if Array 2 were repeated across the first dimension to match the shape of Array 1, making the effective shape of Array 2 (3, 4). 


```python
# some exaples of broadcasting 
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(4)
print(f"array 1 \n:{a2}")
print(f"array 2 \n: {a3}")
```

```python
print(a2 + a3)

# lets get some more of an example of broadcasting 
a4 = np.arange (12).reshape (3,4)
a5 = np.arange(3)
# in this case the broadcasting is not possible because the array a5 have the dimension of 3 then it is recognozed as an (1,3) but the array a4 is in 
# the dimension of (3,4) so the 1D array will result in (3,3) and the broadcasting of the (3,4) and (3,3) is not possoble 
```

```python
# so in the example the 1D array which is 4 is converted to (1,4) shape and the 1 stretches to its corresponding array which is of the shape (3,4) so 
# here 1 == 3 and the array of shape 4 is converted into (3,4) and the operatrion is done 
```

```python
# calculating sigmoid of an numpy array 
arrayy = np.arange(30)
def sigmoid(arrayy):
    return 1/(1+np.exp(arrayy))
sigmoid(arrayy)
```

```python
# mean squared error 
# Mean Squared Error (MSE) is a way to measure how well a model's predictions match the actual data. It calculates the average of the squared differences
# between the predicted values and the actual values. 
```

formua of the mean squared error :
The formula for the mean squared error is MSE = Σ(yi − pi)2n, where yi is the ith observed value, pi is the corresponding predicted value for yi, and n is the number of observations. 
The Σ indicates that a summation is performed over all values of i.

```python
av = np.random.randint(1,50,25)
pv = np.random.randint(1,50,25)
def mse (actual_value,predicted_value):
    return np.mean((actual_value - predicted_value)**2)
mse(av,pv)

# the output is the mean squared error 
```
